```python
#Import the libraries

import tweepy #py lib for twitter API 
from tweepy import OAuthHandler  #authentication for twitter API
from tweepy import Stream
from tweepy.streaming import StreamListener
import json
```

```python
#Credentials for accessing Twitter API
#Making it configurable - TBD;

consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''
```

```python
class TweetsListener(StreamListener) :
    
    #Print twitter-stream to stdout
    def on_data(self, data):
        try:
            msg = json.loads(data)
            if msg['truncated'] is True:
                print(msg['extended_tweet']['full_text'])
            else:
                print(msg['text'])
            return True
        except BaseException as e:
            print("Error : %s" %str(e))
        return True
    
    #Deal with errors - printing on console 
    def on_error(self, status):
        print(status)
        return True

#With valid credentials, start a filtered stream
def sendData():
    auth = OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_secret)
    
    twitter_stream = Stream(auth, TweetsListener(), tweet_mode='extended')
    
    #Making filter string configurable - TBD;
    twitter_stream.filter(track=['trump'])
```

```python
if __name__ == "__main__":
    
    #Generate Twitter Stream 
    sendData()

    #Send data to Kafka-topic ; TBD
```
